```python
import pandas as pd
import matplotlib.pyplot as plt

# 엑셀 파일에서 데이터 읽어오기
file_path = "C:/python/South_GRDP.xlsx"  # 엑셀 파일 경로
df = pd.read_excel(file_path, engine='openpyxl')

# 데이터프레임 확인
print(df)
```

```python

# 막대 그래프 생성
fig, ax = plt.subplots(figsize=(12, 8))

# 막대 그래프 그리기
bars = plt.bar(df["지역"], df["GRDP"], color='b')

# 그래프 제목과 레이블 설정
plt.xlabel('City', fontsize=14)
plt.ylabel('GRDP', fontsize=14)
plt.title('GRDP by City in Gyeonggi Province', fontsize=16)
plt.xticks(rotation=45, ha='right')

# 막대 위에 GRDP 값 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, '%d' % int(height), ha='center', va='bottom')

# 그래프 표시
plt.tight_layout()
plt.show()
layout()
plt.show()
```
