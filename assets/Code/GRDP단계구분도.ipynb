```python
import geopandas as gpd
import matplotlib.pyplot as plt


file_path = './re_merged_data.shp'
data = gpd.read_file(file_path)

print(data.head())
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
data.plot(column='GRDP', cmap='OrRd', legend=True, legend_kwds={'label': "Level of Data"}, ax=ax)



#
plt.show()
```
