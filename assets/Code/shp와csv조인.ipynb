```python
import geopandas as gpd
import pandas as pd

shp_path = './경기도4326.shp'
shp_data = gpd.read_file(shp_path)

xlsx_path = './경기도GRDP.xlsx'
xlsx_data = pd.read_excel(xlsx_path)

shp_data['Modified_SGG_NM'] = shp_data['SGG_NM'].apply(lambda x: x[3:])
```

```python
merged_data = gpd.sjoin(shp_data, xlsx_data, how='inner', op='intersects', left_on='Modified_SGG_NM', right_on='Region')
```

```python
# 데이터 오류가 계속 있어서 공백 제거를 진행함

shp_data['Modified_SGG_NM'] = shp_data['Modified_SGG_NM'].str.strip().str.upper()
xlsx_data['Region'] = xlsx_data['Region'].str.strip().str.upper()
```

```python
merged_data = pd.merge(shp_data, xlsx_data, left_on='Modified_SGG_NM', right_on='Region', how='inner')

# 결과 확인
print(merged_data)
```
