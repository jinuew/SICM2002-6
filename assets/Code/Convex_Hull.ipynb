```python
import geopandas as gpd
from shapely.geometry import MultiPoint
import matplotlib.pyplot as plt

gdf = gpd.read_file('./4326.shp')

fig, ax = plt.subplots(figsize=(10,10))
gdf.plot(ax=ax)

plt.show()
```

```python
try:
    gdf = gpd.read_file('./4326.shp')
    gdf['convex'] = gdf.geometry.apply(lambda x: x.convex_hull)

    convex_hull_gdf = gdf.set_geometry('convex')
  
    # 
    fig, ax = plt.subplots(1, 1, figsize=(20, 10))
    convex_hull_gdf.plot(ax=ax, color='green', edgecolor='k')
    ax.set_title('Convex Hulls')
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])
    plt.show()


except Exception as e:
    error_message = str(e)
    error_message
```
