```python
# 네이버 뉴스 검색 크롤링

import requests
from bs4 import BeautifulSoup as bs
import pandas as pd



# 사용자 검색어 입력
keyword = input("검색어를 입력하세요: ")
# 검색 결과 페이지 수 입력
total_page = int(input("몇 페이지까지 검색할까요: "))


data = []

page_num = 1
for i in range(1, total_page * 10, 10):
    search_url = f"https://search.naver.com/search.naver?where=news&query={keyword}&start={i}"
    response = requests.get(search_url)
    html_text = response.text

    print()
    print(response.status_code)
    print(f'******* {page_num}페이지 검색 결과입니다. *******')

    # 다루기 어려운 html 문자열을 html 태그와 데이터를 추출하기 쉬운 soup객체로 만들어 준다.
    soup = bs(html_text, 'html.parser')

    media_info = soup.select('a.info.press')
    article_info = soup.select('a.news_tit')

    for j in range(len(media_info)):
        media_name = media_info[j].text.split()[0]
        media_website = media_info[j].attrs['href']
        article_title = article_info[j].text
        article_link = article_info[j].attrs['href']
        
        # 각 기사 정보를 data 리스트에 추가
        data.append({
            '언론사': media_name[:-3],
            '홈페이지': media_website,
            '기사 제목': article_title,
            '기사 링크': article_link
        })
    
    page_num += 1

# 데이터 프레임 생성
df = pd.DataFrame(data)
print(df)
```
